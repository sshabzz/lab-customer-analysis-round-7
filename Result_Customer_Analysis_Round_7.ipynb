### ***Using OneHotEncoder to turn all the categorical values into columns***

```python
from sklearn.preprocessing import OneHotEncoder

# Select only the categorical columns from MCA1
categorical_columns = MCA1.select_dtypes(include=['object'])

# Fit the OneHotEncoder to the categorical columns
encoder = OneHotEncoder(drop='first').fit(categorical_columns)

# Transform the categorical columns using the fitted encoder
encoded = encoder.transform(categorical_columns).toarray()

# Create a DataFrame of the one-hot encoded data
onehot_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())

# Combine the one-hot encoded data with the numerical data
X = pd.concat([onehot_encoded, MCA1.select_dtypes(include=['float64', 'int64'])], axis=1)

```

```python
from sklearn.preprocessing import StandardScaler

# Select only the numerical columns from MCA1
numerical_columns = MCA1.select_dtypes(include=['float64', 'int64'])

# Dropping target column
numerical_columns.drop('total_claim_amount', axis=1, inplace=True)

# Normalize the numerical columns using StandardScaler
scaler = StandardScaler().fit(numerical_columns)
x_normalized = scaler.transform(numerical_columns)

# Create a DataFrame of the normalized numerical data
normalized_df = pd.DataFrame(x_normalized, columns=numerical_columns.columns)

# Combine the normalized numerical data with the one-hot encoded categorical data
X = pd.concat([onehot_encoded, normalized_df], axis=1)
```

```python
# Select only the categorical columns from MCA1
categorical_columns = MCA1.select_dtypes(include=['object'])

# Fit the OneHotEncoder to the categorical columns
encoder = OneHotEncoder(drop='first').fit(categorical_columns)

# Transform the categorical columns using the fitted encoder
encoded = encoder.transform(categorical_columns).toarray()

# Create a DataFrame of the one-hot encoded data
onehot_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())

# Combine the one-hot encoded data with the numerical data
X = pd.concat([onehot_encoded, MCA1.select_dtypes(include=['float64', 'int64'])], axis=1)

# Dropping target colum from X
X.drop('total_claim_amount', axis=1, inplace=True)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### ***Apply linear regression***

```python
from sklearn import linear_model

lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
# checking the model predictions on the train and test set. It performes sligtly better on the train set
from sklearn.metrics import r2_score
y_pred_train = lm.predict(X_train)
print("R2 of a train set:", r2_score(y_train, y_pred_train))
y_pred_test = lm.predict(X_test)
print("R2 of a test set:", r2_score(y_test, y_pred_test))
```

### ***Model Validation***
